# 🧹 Data Cleaning & Preprocessing

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
```

```python
df['product_category_name_english'] = df['product_category_name_english'].fillna('Unknown')
```

```python
df['payment_value'] = df['payment_value'].fillna(np.nan)
```

```python
df['payment_value'] = df['payment_value'].fillna(df['payment_value'].median())
```

```python
df['review_score'] = df['review_score'].fillna(df['review_score'].median())
```

```python
print(df.isna().sum())
```

```python
df['order_month'] = df['order_month'].dt.to_timestamp()
```

```python
# Check for duplicate rows
duplicates_count = df.duplicated().sum()
print(f'🔍 Number of Duplicate Rows: {duplicates_count:}')
```

```python
# Check for duplicate Columns
duplicates_count = df.T.duplicated().sum()
print(f'🔍 Number of Duplicate Columns: {duplicates_count:}')
```

```python
df = df.drop_duplicates()
```

```python
# Check for duplicate rows
duplicates_count = df.duplicated().sum()
print(f'🔍 Number of Duplicate Rows: {duplicates_count:}')
```

```python
print(df.isna().sum())
```

# 🧑‍💻 KPI Exploration 

## 1. What is the total revenue generated?


```python
total_revenue = df['payment_value'].sum()
print(f'Total Revenue : {total_revenue}')
```

## 2. How does revenue change over time? Identify trends by month. 

```python
monthly = df.groupby('order_month')['payment_value'].sum()
monthly 
```

```python
monthly['year'] = monthly['order_month'].dt.year
monthly['month'] = monthly['order_month'].dt.month

for year in monthly['year'].unique():
    data = monthly[monthly['year'] == year]
    plt.plot(data['month'], data['payment_value'], marker='o', label=str(year))

plt.title('Monthly Revenue per Year')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

## 3. Which product categories generate the most revenue? 

```python
top_5 = df.groupby('product_category_name_english')['payment_value'].sum().nlargest(5)

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(top_5.index, top_5.values)

ax.bar_label(bars, fmt='{:,.0f}', padding=3)

plt.title('Top 5 Product Categories by Revenue')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.show()
```

## 4. Which categories or products perform the worst?

```python
bottom_5 = df.groupby('product_category_name_english')['payment_value'].sum().nsmallest(5)

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(bottom_5.index, bottom_5.values)

ax.bar_label(bars, fmt='{:,.0f}', padding=3)

plt.title('Bottom 5 Product Categories by Revenue')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.xticks(rotation=45)

plt.show()
```

## 5.Which customer locations (city/state) generate the most orders? 

```python
most_order = df.groupby('customer_city')['customer_city'].count().nlargest(5)

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(most_order.index, most_order.values)

ax.bar_label(bars, fmt='{:,.0f}', padding=3)

plt.title('Top 5 Location by number of orders')
plt.xlabel('Location')
plt.ylabel('Number of orders')

plt.show()
```

## 6. What is the average order value?

```python
average_order_value = df['payment_value'].mean()
print(f' average order value : {average_order_value:}')
```

## 7. How many items are typically included in an order?

```python
average_item_per_order = df['items_per_order'].mean()
print(f' average item per order : {average_item_per_order:}')
```

## 8. What is the relationship between review score and sales or order value?


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='review_score', y='payment_value', data=df, showfliers=False)
plt.title('Distribution of Payment Value per Review Score')
plt.xticks(rotation=45)

plt.show()
```

## 9. Are there any patterns in customer satisfaction?

```python
best_category = df.groupby('product_category_name_english')['review_score'].mean().nlargest(5)
best_category
```

```python
worst_category = df.groupby('product_category_name_english')['review_score'].mean().nsmallest(5)
worst_category
```
